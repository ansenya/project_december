```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

conn = sqlite3.connect('switrs.sqlite')
```

#### This database has 4 tables and many fields in each table, so we have to load the data into one df.

```python
case_ids_df = pd.read_sql_query('SELECT * FROM case_ids', conn)
case_ids_df.head()
```

```python
collisions_df = pd.read_sql_query('SELECT * FROM collisions LIMIT 1000', conn)
collisions_df.head()
```

```python
parties_df = pd.read_sql_query('SELECT * FROM parties LIMIT 1000', conn)
parties_df.head()
```

```python
victims_df = pd.read_sql_query('SELECT * FROM victims LIMIT 1000', conn)
victims_df.head()
```

#### Let's firstly watch how much there are old vehicles collisions then the new one

```python
cars_age_df = pd.read_sql_query('SELECT vehicle_year FROM parties;', conn)
invalid_years_df = cars_age_df[(cars_age_df['vehicle_year'] > 2022) | (cars_age_df['vehicle_year'] < 1900)]

missing_values = cars_age_df.isna().sum()
invalid_years = invalid_years_df.shape[0]

info_df = pd.DataFrame({
    'Column': ['vehicle_year'],
    'Missing Values': [missing_values['vehicle_year']],
    'Invalid Years': [invalid_years],
})
info_df.head()
```

### cleaning

```python
cars_age_df = cars_age_df[
    (cars_age_df['vehicle_year'].notna()) &
    (cars_age_df['vehicle_year'] >= 1980) &
    (cars_age_df['vehicle_year'] <= 2022)
    ]
cars_age_df.head()
```

```python
car_counts = cars_age_df['vehicle_year'].value_counts().sort_index()

plt.figure(figsize=(14, 6))
sns.barplot(x=car_counts.index, y=car_counts.values, color='skyblue')
plt.title('Number of Cars by Manufacture Year Involved in Accidents')
plt.xlabel('Car Manufacture Year')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### as we can see, vehicles made in 2000 collide the most
#### that's probably because there are more such cars, then any others

```python
people_info_df = pd.read_sql_query(
    'SELECT party_sex, party_age, cellphone_in_use, party_race FROM parties where at_fault == 1;', conn)
people_info_df = people_info_df[
    ((people_info_df['party_sex'] == 'female') | (people_info_df['party_sex'] == 'male')) &
    (people_info_df['party_age'].notna()) &
    (people_info_df['cellphone_in_use'].notna()) &
    (people_info_df['party_race'].notna())
    ]
people_info_df.head()
```

```python
male_count = people_info_df[people_info_df['party_sex'] == 'male'].shape[0]
female_count = people_info_df[people_info_df['party_sex'] == 'female'].shape[0]

plt.pie([male_count, female_count], labels=['male', 'female'], autopct='%1.1f%%', colors=['blue', 'pink'])
plt.title('Male vs Female Percentage')
plt.show()
```
